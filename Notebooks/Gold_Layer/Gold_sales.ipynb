```python
query = """
        SELECT 
        order_number,
        dp.product_key,
        dc.customer_key,
        order_date,
        shipping_date,
        due_date,
        sales_amount,
        quantity,
        price 
		FROM silver.crm_sales sd
		LEFT JOIN gold.dim_products dp
		ON sd.product_key = dp.product_number
		LEFT JOIN gold.dim_customers dc
		ON sd.customer_id = dc.customer_id;
"""
df = spark.sql(query)
```

```python
(
    df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("gold.fact_sales")
)
```
