# Read from CSV and Write bronze table

```python
for config in INGESTION_CONFIG:
    print(f"Ingesting {config["source"]} -> bronze.{config["table"]}")
    df = (
        spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(config["path"])
    )
    df.write.mode("overwrite").format("delta").saveAsTable(f"bronze.{config["table"]}")
```
