# Initialization

```python
ERP_PRODUCT_CATEGORY_RENAME_MAP = {
    "id": "category_id",
    "cat": "category",
    "subcat": "subcategory",
    "maintenance": "maintenance"
}
```

# Reading from bronze table

```python
df = spark.table("workspace.bronze.erp_px_cat_g1v2")
```

# Transformations

## Renaming columns

```python
for old_name, new_name in ERP_PRODUCT_CATEGORY_RENAME_MAP.items():
    df = df.withColumnRenamed(old_name, new_name)
```

## Selecting the columns

```python
df = df.select(
    "category_id",
    "category",
    "subcategory",
    "maintenance"
)
```

# Writing to Silver Table

```python
(
    df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("silver.erp_product_category")
)
```
